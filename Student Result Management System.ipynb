**Adding and Calculating Grades for Each Student**

```python
# Automatic grade calculating function
def calculate_grade(score):
    if score >= 70:
      return "A"
    elif score >= 60:
      return "B"
    elif score >= 50:
      return "C"
    elif score >= 45:
      return "D"
    else:
      return "F"
```

**Storing student names and records**

```python
# This is a function for the dictionary that stores the student records
student_records = {}

def add_student():
  try:
    name = str(input("Enter the student's name    "))
     # Check that the name contains only letters
    if not name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    score = float(input("Enter the Student's Score   "))
    if score < 0 or score > 100:
        print("Entry is out of range(0, 100)")
    else:
        print("The student's record has been successfully created")

    student_records[name] = {
        "score": score,
        "grade": calculate_grade(score)
        }

  except ValueError as e:
        print("Error:", e)
```

**Search for student record**

```python
# function to search for students
def search_student():
  try:
    name = str(input("Enter Student's name to search for     "))
    # check that the name contain only letters
    if not name.replace(" ", "").isalpha():
        raise ValueError("Name must contain only letters.")

    # to check if the stundent record exist
    if name not in student_records:
        raise KeyError("Record not found")

    if name in student_records:
      print(f"{name} found with score: {student_records[name]}")

  except ValueError as e:
      print("Error:", e)
  except KeyError as e:
      print("Error:", e)
```

**View all Students**

```python
 # To create a display function
def view_students():
  for name, data in student_records.items():
    print(f"Name: {name},  "
          f"score: {data["score"]},  "
          f"grade: {data["grade"]}.  "
    )
```

```python
view_students() # to check the function
```

**Class Performance Statistics**

```python
# class performance statistics function
def class_statistics():
  scores = []

  for data in student_records.values():
    scores.append(data["score"])

  Total = sum(scores)
  Average = Total/len(scores)
  Highest = max(scores)
  Lowest = min(scores)

  print("Class Performance Statistics")
  print("Total Score:", Total)
  print("Average Score:", Average)
  print("Highest Score:", Highest)
  print("Lowest Score:", Lowest)
```

```python
class_statistics() # To display the Class Performance Statistics
```
